```python
import torch

train_dict = torch.load('dataset/vehicle_dataset_ext_train.pt')
test_dict = torch.load('dataset/vehicle_dataset_ext_test.pt')

print(f"Train features shape: {train_dict['features'].shape}")
print(f"Train targets shape: {train_dict['targets'].shape}")
print(f"Test features shape: {test_dict['features'].shape}")
print(f"Test targets shape: {test_dict['targets'].shape}")



test_targets_overlap = 0
for test_target in test_dict['targets']:
  if any(torch.isin(train_dict['targets'], test_target)):
    # print("Found overlapping target:", test_target)
    test_targets_overlap += 1
print(f"Number of overlapping targets in test set: {test_targets_overlap} out of {len(test_dict['targets'])}")



test_features = test_dict['features'].tolist()
train_features = train_dict['features'].tolist()

```

Check for target overlap:

```python
test_targets_overlap = 0
for test_target in test_dict['targets']:
  if any(torch.isin(train_dict['targets'], test_target)):
    # print("Found overlapping target:", test_target)
    test_targets_overlap += 1
print(f"Number of overlapping targets in test set: {test_targets_overlap} out of {len(test_dict['targets'])}")
```

Check for feature overlap:

```python

test_features_overlap = 0


# Compare the first 2 features (index 0 and 1) from each feature vector
test_first_feats = [tuple(f[0:2]) for f in test_features]
train_first_feats = [tuple(f[0:2]) for f in train_features]
test_features_overlap = 0

print(test_first_feats[0:4])
for test_feature in test_first_feats:
  if test_feature in train_first_feats:
    # Found overlapping first-2-features value
    test_features_overlap += 1

print(f"Number of overlapping first-2-feature values in test set: {test_features_overlap} out of {len(test_dict['features'])}")

print(f"Number of overlapping features in test set: {test_features_overlap} out of {len(test_dict['features'])}")
```
